```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
Bcancer_dataset= sklearn.datasets.load_breast_cancer()
```

```python
print(Bcancer_dataset)
```

```python
df=pd.DataFrame(Bcancer_dataset.data, columns=Bcancer_dataset.feature_names)
```

```python
df['label']= Bcancer_dataset.target
```

```python
X= df.drop(columns='label', axis=1)
Y=df['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train,X_test,Y_train, Y_test = train_test_split(X,Y, test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

MODEL TRAINING


```python
lr=LogisticRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
X_predict=lr.predict(X_test)
```

```python
y_predict=accuracy_score(Y_test,X_predict)
```

```python
print(y_predict)
```

DEEP LEARNING MODEL

```python
from sklearn.preprocessing import StandardScaler
```

```python
scl=StandardScaler()
```

```python
X_train_s = scl.fit_transform(X_train)
X_test_s =scl.transform(X_test)
```

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(30,)),
                          keras.layers.Dense(20, activation='relu'),
                          keras.layers.Dense(2,activation='sigmoid')

])
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history= model.fit(X_train_s, Y_train, validation_split=0.1, epochs=10)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training data','validation data'], loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['training data','validation data'], loc='upper right')
```

```python
loss,accuracy = model.evaluate(X_test_s, Y_test)
print(accuracy)
```
